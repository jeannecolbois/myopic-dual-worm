```python
xdata = 1./(3*np.array(L))
ydata = np.array(S)[:,0]
vSmax = np.array(Smax)[:,0]
vSmin = np.array(Smin)[:,0]
sigma = (vSmax - vSmin)/2
```

```python
def poly(x, b, c):
    return b*x**2 + c
```

```python
idmin = 3
```

```python
[popt, perr] = curve_fit(poly, xdata[idmin:], ydata[idmin:], sigma = sigma[idmin:], absolute_sigma = True)
```
