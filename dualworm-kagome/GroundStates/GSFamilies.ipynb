```python
kw = {'binning': False, 'plzplot': True, 'plotmin': 0, 'plotmax': 16}

[groundspinstates, groundstates, groundcharges] = \
 af.LoadGroundStates(foldernames[0], filenamelist, L, nh, [nsm],**kw)

groundspinstates = groundspinstates[0]
groundstates = groundstates[0]
groundcharges = groundcharges[0]

for i in range(1,len(foldernames)):
    [gsps, gs, gc] = \
     af.LoadGroundStates(foldernames[i], filenamelist, L, nh, [nsm],**kw)
    groundspinstates = np.concatenate((groundspinstates, gsps[0]))
    groundstates = np.concatenate((groundstates, gs[0]))
    groundcharges = np.concatenate((groundcharges, gc[0]))
    
[d_ijl, ijl_d, s_ijl, ijl_s, d_2s, s2_d, d_nd, d_vd, d_wn,
 sidlist, didlist, c_ijl, ijl_c, c2s, csign] =\
dw.latticeinit(L[0])
(s_ijl, ijl_s) = kdraw.createspinsitetable(L[0])


backup = "./"+foldernames[0]+filenamelist[0]+".hkl"
couplings = hkl.load(backup, path="/parameters/couplings")
hamiltonian = dw.Hamiltonian(couplings, d_ijl, ijl_d, L[0])

if type == 'J1J2':
    gsenergy = -2/3 * J1[0] - 2/3 * J2[0]
elif type == 'J1J3':
    gsenergy = -2/3 * J1[0] -  J3[0]
elif type == 'inter':
    gsenergy = -2/3 * J1[0] - 1/3 * J3[0]
elif type == 'LargeJ3':
    gsenergy = -2/3 * J1[0] + 2/3 * J2[0] -  J3[0]
latsize = len(s_ijl)
families, spinfamilies = mfu.FamiliesFromStates(hamiltonian, groundstates,
                                      gsenergy, groundspinstates,
                                      d_wn, latsize, 0.5, d_ijl)
```
